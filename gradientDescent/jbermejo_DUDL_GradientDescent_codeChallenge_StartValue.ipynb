```python
import numpy as np
import sympy as sym

import matplotlib.pyplot as plt
```

# Step 1: repeat 1 dimension gradient descent for the following function:
f(x) = cos(2pi*x) + x^2, x â‚¬ {-2, 2}

```python
# We create the expresion of the functions
sym_x = sym.symbols('x')
fx_expr = sym.cos(2 * sym.pi * sym_x) + sym_x**2
df_expr = sym.diff(fx_expr, sym_x)

print(fx_expr)
print(df_expr)
```

```python

# We copy and transform the expressions into functions that work with numpy
def fx(x):
    return x**2 + np.cos(2 * np.pi * x)

def df(x):
    return 2*x - 2*np.pi*np.sin(2*np.pi*x)
```

```python
# Plotting the functions

# We define the range of the plot according to definition
x = np.linspace(-2, 2, 100)

# we plot
plt.plot(x, fx(x), x, df(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)', 'f\'(x)'])
plt.show()
```

```python
# random starting point 
localmin = np.random.uniform(-2, 2)
print(f'x0 = {localmin}')

# learning parameters
learning_rate = 0.01
training_epochs = 10000

# run the gradient descent algorithm
for epoch in range(training_epochs):
    grad = df(localmin)
    localmin = localmin - learning_rate*grad

print(localmin)
```

```python
# plot the results

plt.plot(x, fx(x), x, df(x))
plt.plot(localmin, df(localmin), 'ro')
plt.plot(localmin, fx(localmin), 'ro')

plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)', 'f\'(x)'])
plt.title('Empirical local minimum: %x'%int(localmin))
plt.show()
```
