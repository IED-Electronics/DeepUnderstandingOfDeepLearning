```python
# import libraries
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
# create data

N = 150
x = torch.randn(N,1)
y = x + torch.randn(N,1)/2

# and plot
plt.plot(x,y,'s')
plt.show()
```

```python
# build model
ANNreg = nn.Sequential(
    nn.Linear(1,1,),  # input layer
    nn.ReLU(),       # activation function
    nn.Linear(1,1)   # output layer
    )

ANNreg
```

```python
# learning rate
learningRate = .05

# loss function
lossfun = nn.MSELoss()

# optimizer (the flavor of gradient descent to implement)
# basically, we add the gradient descent to the model (SGD = Stochastic Gradient Descent)
optimizer = torch.optim.SGD(ANNreg.parameters(),lr=learningRate)
```

```python
# train the model
numepochs = 500
losses = torch.zeros(numepochs)


## Train the model!
for epochi in range(numepochs):

  # forward pass 
  # we take the observed data (x) and we pass it through the model (to get the prediction)
  yHat = ANNreg(x)

  # compute loss (using the MSE loss function that we created previously)
  # then we add it to the losses array
  loss = lossfun(yHat,y)
  losses[epochi] = loss

  # backprop
  # we do the backpropagation, that is the gradient descent
  optimizer.zero_grad() # we initialize the derivatives of the gradients to zero
  loss.backward() # we compute the gradients (backpropagation per se)
  optimizer.step() # housekeeping of the optimizer of the gradient 
```

```python
# show the losses

# manually compute losses
# final forward pass (in other words, final prediction)
predictions = ANNreg(x)

# final loss (MSE) (we do it manually instead of using lossfun)
testloss = (predictions-y).pow(2).mean()

plt.plot(losses.detach(),'o',markerfacecolor='w',linewidth=.1)
plt.plot(numepochs,testloss.detach(),'ro')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Final loss = %g' %testloss.item())
plt.show()
```

```python
print(testloss)
print(testloss.detach())
print(testloss.item())
```

```python
# plot the data
plt.plot(x,y,'bo',label='Real data')
plt.plot(x,predictions.detach(),'rs',label='Predictions')
plt.title(f'prediction-data r={np.corrcoef(y.T,predictions.detach().T)[0,1]:.2f}')
plt.legend()
plt.show()
```
